```python
import pandas as pd
import sqlite3
```

```python
df_sales = pd.read_csv("dubai_clean_sales_price.csv")
df_sales
```

```python
if "Unnamed: 0" in df_sales.columns:
    del df_sales["Unnamed: 0"]
df_sales.index = range(1, len(df_sales)+1)
df_sales
```

```python
#Add corresponding year column make it easier to aggregate.
df_sales['instance_date'] = pd.to_datetime(df_sales['instance_date'], errors='coerce') # Convert to datetime if not already
df_sales["year"] = df_sales["instance_date"].dt.year
cols = df_sales.columns.tolist() 
cols.insert(1,cols.pop(cols.index("year")))
df_sales = df_sales[cols]
df_sales
```

```python
df_sales['instance_date'] = pd.to_datetime(df_sales['instance_date'], errors='coerce')
print(df_sales['instance_date'].dtype)
df_sales['instance_date'] = df_sales['instance_date'].dt.strftime('%m/%d/%Y')
df_sales
```

```python
with sqlite3.connect("sales.db") as conn:
    df_sales.to_sql("sales_data", conn, if_exists="replace", index=False)
    query1 = """
        SELECT
            year,
            property_sub_type_en,
            AVG(actual_worth) AS average_sales_price
        FROM
            sales_data
        GROUP BY
            year, property_sub_type_en   
    """
    res1 = pd.read_sql_query(query1,conn)
res1
```

```python
query2 = """
        SELECT
            year,
            area_name_en,
            AVG(actual_worth) AS average_sales_price
        FROM
            sales_data
        GROUP BY
            year, area_name_en  
    """
res2 = pd.read_sql_query(query2, conn)
res2
```
